#K Means clustering

```python
data.columns
```

```python
x = data[['Channel', 'Region','Fresh', 'Milk', 'Grocery', 'Frozen','Detergents_Paper', 'Delicassen']]
```

```python
from sklearn.cluster import KMeans
```

Finding the wcss score

```python
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters = i, n_init = 10)
  km.fit(x)
  css = km.inertia_
  wcss.append(css)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss,'*-')
plt.xlabel('No of clusters')
plt.ylabel('wcss')
plt.show()
```

From elbow mathod 3 is the optimum number of clusters

Creating th model with 3 clusters





```python
 km = KMeans(n_clusters = 3, n_init = 10)
```

```python
km.fit(x)
```

```python
y_km = km.predict(x)
```

```python
y_km
```

Finding the number of datapoints in each cluster

```python
pd.Series(y_km).value_counts()
```

Added a new column 'Cluster' to show data points in different clusters

```python
data['Cluster'] = y_km
```

```python
data.head()
```

Finding the Silhouette score





```python
from sklearn.metrics import silhouette_score
silhoutte = silhouette_score(x,y_km)
print(silhoutte)
```

#**Agglomerative HC**

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

Finding the distance between the centroids

```python
mergings = linkage(x,method = 'centroid')
```

```python
plt.figure(figsize=(10,8))
dendrogram(mergings)
plt.show()
```

Using Ward method

```python
merging = linkage(x,method = 'ward')
```

```python
plt.figure(figsize=(12,8))
dendrogram(merging)
plt.axhline(150000,color = 'Black',linestyle ='--')
plt.grid()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, metric = 'euclidean', linkage ='ward')
y_hc = hc.fit_predict(x)
```

```python
y_hc
```

Finding the Silhouette score

```python
from sklearn.metrics import silhouette_score
silhoutte = silhouette_score(x,y_km)
print(silhoutte)
```
